```python
#codigo normal para hacer problema de fibonnaci pero no calcula grandes numeros como de 50 tarda mucho 

def fibonacci_recursivo(n):
    if n == 0 or n == 1:
        return 1

    return fibonacci_recursivo(n - 1) + fibonacci_recursivo(n - 2)
if __name__ == '__main__':
    n = int (input ('escoge un numero '))
    resultado = fibonacci_recursivo(n)
    print(resultado)
```

```python
# hacerlo de manera mas eficiente para calcular numeros mas grandes pero a numeros mas grandes no calcula esos numeros 

def fibonacci_recursivo(n):
    if n == 1 or n == 1:
        return 1
    return fibonacci_recursivo(n-1)+fibonacci_recursivo(n-2)
def fibonacci_dinamico(n, memo = {}):
    if n == 0 or n == 1:
        return 1

    try:
        return memo[n]
    except KeyError: # pedir permiso y estar acceder al numero  
        resultado = fibonacci_dinamico(n - 1, memo) + fibonacci_dinamico(n - 2, memo)
        memo[n] = resultado

        return resultado

        
if __name__ == '__main__':
    n = int (input ('escoge un numero '))
    resultado = fibonacci_dinamico(n)
    print(resultado)
```

```python
# evitar error de  maxima recursion  ecede la comparacion 
# se resulve con  import sys 
import sys

def fibonacci_recursivo(n):
    if n == 0 or n == 1:
        return 1

    return fibonacci_recursivo(n - 1) + fibonacci_recursivo(n - 2)


def fibonacci_dinamico(n, memo = {}):
    if n == 0 or n == 1:
        return 1

    try:
        return memo[n]
    except KeyError:
        resultado = fibonacci_dinamico(n - 1, memo) + fibonacci_dinamico(n - 2, memo)
        memo[n] = resultado

        return resultado

if __name__ == '__main__':
    sys.setrecursionlimit(10002)
    n = int(input('Escoge un numero: '))
    resultado = fibonacci_dinamico(n)
    print(resultado)
```
